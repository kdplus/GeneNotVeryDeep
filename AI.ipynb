### Divide data into batch

```python
raw_data = np.load("./raw_data.npy")
raw_label = np.load("./label.npy")
```

```python
index = np.random.permutation(raw_data.shape[0])
```

```python
data = raw_data[index]
label = raw_label[index]
```

```python
train_size = int(data.shape[0] * 0.8)
test_size = data.shape[0] - train_size
```

```python
train_data = data[0:train_size]
train_label = label[0:train_size]
test_data = data[train_size:]
test_label = label[train_size:]
```

```python
np.save('train_data', train_data)
np.save('train_label', train_label)
np.save('test_data', test_data)
np.save('test_label', test_label)
```

```python
def get_batch(data, batch_size, capacity):
    data = tf.cast(data, tf.float32)
    input_queue = tf.train.slice_input_producer([data])
    data = input_queue[0]
    data_batch = tf.train.batch([data], batch_size=batch_size, num_threads=8, 
                                    capacity=300002,allow_smaller_final_batch=False)
    return data_batch
```

```python
BATCH_SIZE = 16
CAPACITY = 30000
one_batch = get_batch(data, BATCH_SIZE, CAPACITY)
```
